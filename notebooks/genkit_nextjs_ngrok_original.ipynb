# Firebase GenKit:

Firebase Genkit, geliştiricilerin yeni ve mevcut uygulamalarına gelişmiş yapay zeka yetenekleri entegre etmelerini kolaylaştırmayı hedeflemektedir.

Genkit, Google'ın güçlü büyük dil modellerini (LLM'ler) daha geniş bir geliştirici kitlesi için erişilebilir kılma stratejisinin bir parçası durumundadır.

Firebase ekosistemine entegre bir şekilde tasarlanan Genkit, tanıdık araçlar ve iş akışları sunarak kullanım kolaylığı sağlıyor. Geliştiriciler, yeni yapay zeka özelliklerini yerel ortamlarında test edebilir ve ardından Firebase için Cloud Functions ve Google Cloud Run gibi Google'ın sunucusuz bilgi işlem platformlarını kullanarak uygulamalarını sorunsuz bir şekilde ölçeklendirebilirler.

Genkit'in açık kaynak yapısı, geliştiricilere genişletme ve özelleştirme konusunda esneklik sağlıyor. Geniş bir üçüncü taraf açık kaynak projesi yelpazesini destekleyen Genkit, Google'ın kendi Gemini modellerinin yanı sıra "Ollama" platformu aracılığıyla açık modellere de erişim sunuyor.

Veri yönetimi açısından da oldukça esnek olan Genkit, Google Cloud Firestore'un yanı sıra Chrome, Pinecone ve PostgreSQL'in pgvector gibi çeşitli vektör veritabanlarını da destekliyor. Ayrıca, eklenti sistemi sayesinde Genkit, farklı modeller, vektör depoları, enbedding modülleri, değerlendiriciler ve diğer bileşenlerle kolayca entegre edilebilir.


**Temel Özellikler:**

* Birçok Model, Tek Arayüz: Genkit, popüler modellere anında erişim sağlayan eklentiler ve herhangi bir model API'sini entegre etmeyi ve topluluk tarafından sürdürülen modelleri kullanmayı kolaylaştıran esnek bir model soyutlaması sağlar.

* Yapılandırılmış Çıktı: Genkit ile Zod şemasını kullanarak güçlü bir şekilde yazılmış veriler oluşturun. Bu, yapılandırılmamış metinleri analiz etmenize, yaratıcı içerik oluşturmanıza, görevleri seçmenize ve sonuçları uygulamanıza yapılandırılmış tür güvenli nesneler olarak geri göndermenize yardımcı olabilir.

* Çok Modlu, Çoklu Ortam: Genkit, metin, veri ve rastgele ortamların karışımlarını destekleyen içerik için ortak bir biçim sağlar. Bu, Genkit'i yalnızca LLM'ler için değil, herhangi bir üretken görevi (görüntü oluşturma gibi) gerçekleştiren modeller için kullanmanıza olanak tanır.

* LLM'lere Araçlar Verin: Genkit, araçlar aracılığıyla LLM'lerle işlev çağrısını basitleştirir. Araçlar, AI'nın veri getirmesini, kullanıcı arayüzü görüntülemesini, bir veritabanına yazmasını veya kodlayabileceğiniz başka herhangi bir eylemi gerçekleştirmesini mümkün kılar.

* Dotprompt ile İstem Yönetimi: İstem mühendisliği, yalnızca metni ince ayar yapmaktan daha fazlasıdır. Kullandığınız model, sağladığınız parametreler ve talep ettiğiniz biçim, çıktınızın kalitesini etkiler. Genkit, daha kolay test ve organizasyon için hepsini tek bir dosyaya koymanıza olanak tanıyan bir istem dosya biçimi olan Dotprompt'u sunar.

* Akışları Yerel Olarak Çalıştırın: Üretken AI, sonuçlarda bir çok varyasyona sahiptir ve bu nedenle denemeler yapmak çok önemlidir. Yerel Genkit Geliştirici Kullanıcı Arayüzü, modeller ve retrievers gibi temel AI bileşenleriyle etkileşim kurmanıza olanak tanır.

* İstediğiniz Yerde Dağıtın: Genkit, modellerinizi ve akışlarınızı Firebase'te veya tercih ettiğiniz başka herhangi bir Node.js platformunda dağıtmak için basit, bildirime dayalı bir yol sunar.

* Üretim İçin Güvenlik ve Ölçeklenebilirlik: Genkit, Firebase'in güvenlik, ölçeklenebilirlik ve güvenilirlik özelliklerinden yararlanır.

* Gözlemlenebilirlik ve İzleme: Genkit, modellerinizin ve akışlarınızın performansını izlemenize ve sorunları hızlı bir şekilde belirleyip gidermenize yardımcı olan yerleşik araçlar sağlar.

* Açık Kaynak ve Genişletilebilir: Genkit, açık kaynaklı bir çerçevedir, yani topluluğun gücünden yararlanabilirsiniz. Genkit'i kendi özel eklentilerinizle de genişletebilirsiniz.

**9.1**-İlk güncellenecek dosyamız "/firstproject/app/genkit.ts" açıyoruz.içeriğini aşağıdaki kod ile değiştiriyoruz. İlk önce gerekli kütüphaneler import edilir. Daha sonra, configureGenkit ayarlanır ve Türk yemeklerinin tariflerini alacağımız promptu ve modeli içeren kod bloğu menuSuggestionFlow ile entegre edilir.

    "use server";

    import * as z from 'zod';

    // Import the Genkit core libraries and plugins.
    import { generate } from '@genkit-ai/ai';
    import { configureGenkit } from '@genkit-ai/core';
    import { defineFlow, startFlowsServer, runFlow } from '@genkit-ai/flow';

    // Import models from the Google AI plugin. The Google AI API provides access to
    // several generative models. Here, we import Gemini 1.5 Flash.
    import {
      googleAI,
      gemini15Flash,
      gemini15Pro,
    } from '@genkit-ai/googleai';

    configureGenkit({
      plugins: [
        googleAI()
      ],
      logLevel: "debug",
      enableTracingAndMetrics: true,
    });

    // Define a simple flow that prompts an LLM to generate turkish food recipes.
    const menuSuggestionFlow = defineFlow(
      {
        name: 'Türk Yemekleri Flow_Gemini 1.5 Flash',
        inputSchema: z.string(),
        outputSchema: z.string(),
      },
      async (subject) => {
        const llmResponse = await generate({
          prompt: `Bana restaurantlarda veya evlerde yapılan ${subject} yemeğinin tarfini yaz.`,
          model: gemini15Flash,
          config: {
            temperature: 1,
          },
        });

        return llmResponse.text();
      }
    );

    export async function callMenuSuggestionFlow(theme: string) {
      const flowResponse = await runFlow(menuSuggestionFlow, theme); // runFlow kullanımı
      console.log(flowResponse);
      return flowResponse;
    }
